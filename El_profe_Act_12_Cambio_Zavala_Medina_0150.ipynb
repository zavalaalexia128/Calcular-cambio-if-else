## Variables
1. CP ---> Costo del producto
2. CD ---> Cantidad de dinero
3. Cambio

## Formulas
$Cambio = CD - CP$

## Diagrama de flujo LucidChart
![Diagrama de flujo](https://raw.githubusercontent.com/zavalaalexia128/Calcular-cambio-if-else/refs/heads/main/DF%20if%20else%20cambio.png)

```python
# Para cuando es verdadero (Si)
# Calculando el cambio usando if else
# datos de entra CD, CP
# mostrar Cambio si es falso
# Si es verdadero mostrar No alcanza
CP = int(input("Ingrese el valor de CP: "))
CD = int(input("Ingrese el valor de CD: "))
if CP >= CD:
    print("No alcanza")
else:
  Cambio = CD - CP
  print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Zavala Medina NL = 54")
```

```python
# Para cuando es falso (No)
# Calculando el cambio usando if else
# datos de entra CD, CP
# mostrar Cambio si es falso
# Si es verdadero mostrar No alcanza
CD = int(input("Ingrese el valor de CD: "))
CP = int(input("Ingrese el valor de CP: "))
if CP >= CD:
    print("No alcanza")
else:
  Cambio = CD - CP
  print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Zavala Medina NL = 54")
```
